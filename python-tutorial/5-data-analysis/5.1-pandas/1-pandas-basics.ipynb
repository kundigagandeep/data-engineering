#### Sorting

```python
#Let's say you wish to see the most expensive cars, you can sort the results using sort_values()
#By default, the data is sorted in an ascending order i.e. low to high, adding argument ascending=False, will results in a descending order.

df.sort_values('Price_in_thousands',ascending=False)
```

#### Group by

```python
df.groupby(by=['Manufacturer'])[['Sales_in_thousands']].sum(numeric_only=False).sort_values(by=['Sales_in_thousands'],ascending=False).reset_index()
```

```python
df.groupby(by=['Manufacturer'])[['Engine_size']].mean().round(2).sort_values(by=['Engine_size'],ascending=False).reset_index()
```

```python
df.groupby(by=['Manufacturer'])[['Fuel_efficiency']].mean().round(2).sort_values(by=['Fuel_efficiency'],ascending=True).reset_index()
```

```python
df.groupby(by=['Manufacturer','Model'])[['Fuel_efficiency']].mean().round(2).sort_values(by=['Fuel_efficiency'],ascending=True).reset_index()
```
