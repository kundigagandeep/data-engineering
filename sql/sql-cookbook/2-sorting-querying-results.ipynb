### ORDER BY <br>
**2.1 Returning Query Results in a Specified Order Problem** <br>
You want to display the names, jobs, and salaries of employees in department 10 in order based on their salary (from lowest to highest).

```python
%load_ext sql
```

```python
%sql postgresql://postgres:Speedy3861@localhost:5432/sql-cookbook
```

```python
%%sql

select *
from employees
where deptno = 10
order by sal asc
```

*Note: When you do order by and column name, it's by default in ascending order, so even if you don't add the keyword 'asc' at the end, the results will be in acsending order.*

```python
%%sql

select *
from employees
where deptno = 10
order by sal
```

*Note: You can also order by using column number as well.*

```python
%%sql

select *
from employees
where deptno = 10
order by 6
```

### ORDER BY DESC

```python
%%sql

select *
from employees
where deptno = 10
order by sal desc
```

### 2.3 Sorting by Substrings

**Problem** <br>
You want to sort the results of a query by specific parts of a string. For example, you want to return employee names and jobs from table EMP and sort by the last three characters in the JOB field. 

```python
%%sql

select ename, job
from employees
order by substring(job,length(job)-2)
```

```python
%%sql

select ename, job, substring(ename,1,1) as first_alphabet, substring(reverse(ename),1,1) last_alphabet
from employees
```
